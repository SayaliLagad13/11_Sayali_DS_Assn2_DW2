```python
df1.shape
```

```python
Q1 = np.percentile(df1['dis'], 25, method='midpoint')
Q3 = np.percentile(df1['dis'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)
```

```python
high = Q3 + (1.5*IQR)
upper_array = np.where(df1['dis'] >= high)
print("Upper Bound:", high)

low = Q1 - (1.5*IQR)
lower_array = np.where(df1['dis'] <= low)
print("Lower Bound:", low)
```

```python
df1.drop(upper_array[0], inplace=True)
df1.drop(lower_array[0], inplace=True)
```

```python
new_shape = df1.shape
print("New Shape:", new_shape)
```

```python
import seaborn as sns
df = pd.read_csv('Iris.csv')
print('Original Dataset')
df.head()
```

```python
df1 = df.drop('Species' , axis=1)
df1_norm = (df1-df1.min())/(df1.max() - df1.min())
df1_norm = pd.concat((df1_norm, df.Species), 1)
print("Scaled Dataset Using Pandas")
df1_norm.head()
```

```python
df1_norm.skew(axis = 0, skipna=True)
```

```python
df.skew(axis = 1, skipna = True)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
```

```python
print("\n\nFor SepalLengthCm")
mean = statistics.mean(df1_norm["SepalLengthCm"])
sd = statistics.stdev(df1_norm["SepalLengthCm"])

plt.plot(df1_norm["SepalLengthCm"], norm.pdf(df1_norm["SepalLengthCm"], mean, sd))
plt.show()
print(mean)
print(sd)

print("\n\nFor PetalWidthCm")
mean = statistics.mean(df1_norm["PetalWidthCm"])
sd = statistics.stdev(df1_norm["PetalWidthCm"])

plt.plot(df1_norm["PetalWidthCm"], norm.pdf(df1_norm["PetalWidthCm"], mean, sd))
plt.show()
print(mean)
print(sd)
```

```python
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
data = pd.read_csv('Iris.csv')
print('Original Dataset')
df = data.drop(columns =['Species','Id'], axis=1)
df.head()

```

```python
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df.to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
print("Scaled Dataset Using MinMaxScaler")
df_scaled.head()
```

```python
mean = statistics.mean(df_scaled["SepalLengthCm"])
sd = statistics.stdev(df_scaled["SepalLengthCm"])

plt.plot(df_scaled["SepalLengthCm"], norm.pdf(df_scaled["SepalLengthCm"], mean, sd))
plt.show()
print(mean)
print(sd)
```

```python
mean = statistics.mean(df_scaled["PetalWidthCm"])
sd = statistics.stdev(df_scaled["PetalWidthCm"])

plt.plot(df_scaled["PetalWidthCm"], norm.pdf(df_scaled["PetalWidthCm"], mean, sd))
plt.show()
print(mean)
print(sd)
```
